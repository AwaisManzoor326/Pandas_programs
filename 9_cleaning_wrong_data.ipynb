```python
import pandas as pd   
```

```python
# wrong data is not the empty cell or wrong format instead it is a wrong entry
# eg. someone wrote 1000 instead of 100
```

```python
# one way to clean wrong data is to replace it with something else
df = pd.read_csv('data.csv')
df.loc[7, 'Duration'] = 4500
print(df.to_string())
# for small datasets we can do it manually but for large datasets, it is very difficult
```

```python
# for large datasets we can define some boundaries or define some rules
# eg. duration should not be more than 120 minutes
for x in df.index:
    if df.loc[x, 'Duration'] > 120:
        df.loc[x, 'Duration'] = 1200
print(df.to_string())
```

```python
# another way to handle wrong data is to remove the entire row
for x in df.index:
    if df.loc[x, 'Duration'] > 100:
        df.drop(x, inplace=True)
print(df.to_string())
```
